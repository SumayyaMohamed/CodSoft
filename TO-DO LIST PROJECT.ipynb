### ********* TO-DO LIST PYHTON PROJECT ****************************

###### In this project, I have implemented a to-do list application using two different programming paradigms: traditional procedural programming in Python and object-oriented programming (OOP) and also implemented by using GUI tkinter library.

### 1. Procedural Implementation:

```python
tasks=[]

while True:
    print("1.Add Task")
    print("2.view task")
    print("3.mark as done")
    print("4.delete task")
    print("5.quit")
    
    
    choice=input("Enter your choice")
    if choice=="1":
        task=input(str("enter the task:"))
        tasks.append(task)
        
        
    elif choice=="2":
        for i,task in enumerate(tasks,1):
            print("{}. {}".format(i, task))
        
    elif choice=="3":
        try:
            index=int(input("Enter the task index to mark as done"))
            tasks[index - 1] += " (Complete)"
        except IndexError:
            print("Invalid task index.Please enter a valid index.")
        except ValueError:
            print("Invalid input .Please enter a valid interger index.")
        
    elif choice=="4":
        index=int(input("Enter the task index to delete:"))
        if 1<= index <= len(tasks):
            del tasks[index-1]
        else:
            print("Invalid task index.Please enter a valid task index.")
    
    
    elif choice=="5":
        break;
        
    else:
        print("invalid number")
```

### 2. Object-Oriented Implementation:

```python
class Todolist():
    #properties and methods
    #init
    def __init__(self):
        self.tasks=[]
        
    def add_task(self,task):
        self.tasks.append(task)
        
    def view_task(self):
        for i,task in enumerate (self.tasks,1):
            print(f"{i}.{task}")
            
    def mark_as_done(self,task_index):
        if 1<= task_index <=len(self.tasks):
            self.tasks[task_index-1]+="(complete)"
            
        else:
            print("Invalid task index.Please Enter a valid index value")
            
    def del_task(self,task_index):
        if 1<= task_index <= len(self.tasks):
            del self.tasks[task_index-1]
            
        else:
            print("Invalid task index .Please Enter a valid index value")
            
    
    
todolist=Todolist()

while True:
    
    print("1.add task")
    print("2.view task")
    print("3.mark as done")
    print("4.delete task")
    print("5.quit")
    
    choice=input("Enter the choice")
    
    if choice=="1":
        task=input("Enter the task:")
        todolist.add_task(task)
        
    elif choice=="2":
        todolist.view_task()
        
    elif choice=="3":
        index=int(input("Enter the task index:"))
        todolist.mark_as_done(index)
    
    elif choice=="4":
        index=int(input("Enter the task index:"))
        todolist.del_task(index)
    
    elif choice=="5":
        break;
        
    else:
        print("Invalid  number")
```

### GUI Implementation


### From Code to Interface: Tkinter To-Do List Application

```python
import tkinter as tk
```

```python
from tkinter import *
from tkinter import messagebox
```

```python
def add_task():
    task=text.get()
    if task:
        todo_list.append(task)
        update_listbox()
        text.set("")
        
    else:
        messagebox.showwarning("Warning","Please Enter a task")
                
def remove_task():
    selected_task=list_box.curselection()
    if selected_task:
        todo_list.pop(selected_task[0])
        update_listbox()
        
    else:
        messagebox.showwarning("Warning","Please selet a task")

def view_task():
    list_box.delete(0,tk.END)
    for task in todo_list:
        list_box.insert(tk.END,task)
        
def mark_as_done():
    selected_task=list_box.curselection()
    if selected_task:
        task_index=selected_task[0]
        todo_list[task_index]+="(complete)"
        update_listbox()
        
    else:
        messagebox.showwarning("Warning","Please select a task")
        
def update_listbox():
    view_task()

root= tk.Tk()
root.title("TO DO List")
root.geometry("500x500")
root.resizable(False,False)
root.config(bg="#E6E6FA")
todo_list=[]


# heading 
title= tk.Label(root,text="TO DO List",font=80)
title.pack(pady=20)
# title.grid(row=3,col=3)


#Entry
text=tk.StringVar()
text_entry=tk.Entry(root,width=35,textvariable=text,font=12)
text_entry.pack()

# buttons
add=tk.Button(root,text="ADD TASK",width=12,command=add_task)
add.place(x=350,y=110)

remove=tk.Button(root,text="REMOVE TASK",width=12,command=remove_task)
remove.place(x=350,y=160)

view=tk.Button(root,text="VIEW TASK",width=12,command=view_task)
view.place(x=350,y=210)

mark_as_done=tk.Button(root,text="MARK AS DONE", width=12,command=mark_as_done)
mark_as_done.place(x=350,y=260)

#list box
list_box=tk.Listbox(root,height=15,width=25,font=12)
list_box.place(x=100,y=100)


root.mainloop()
```
