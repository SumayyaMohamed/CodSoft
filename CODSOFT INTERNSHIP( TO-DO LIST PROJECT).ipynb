### ********* TO-DO LIST PYHTON PROJECT ****************************

###### In this project, I have implemented a to-do list application using two different programming paradigms: traditional procedural programming in Python and object-oriented programming (OOP).

1. Procedural Implementation:

```python
tasks=[]

while True:
    print("1.Add Task")
    print("2.view task")
    print("3.mark as done")
    print("4.delete task")
    print("5.quit")
    
    
    choice=input("Enter your choice")
    if choice=="1":
        task=input(str("enter the task:"))
        tasks.append(task)
        
        
    elif choice=="2":
        for i,task in enumerate(tasks,1):
            print("{}. {}".format(i, task))
        
    elif choice=="3":
        try:
            index=int(input("Enter the task index to mark as done"))
            tasks[index - 1] += " (Complete)"
        except IndexError:
            print("Invalid task index.Please enter a valid index.")
        except ValueError:
            print("Invalid input .Please enter a valid interger index.")
        
    elif choice=="4":
        index=int(input("Enter the task index to delete:"))
        if 1<= index <= len(tasks):
            del tasks[index-1]
        else:
            print("Invalid task index.Please enter a valid task index.")
    
    
    elif choice=="5":
        break;
        
    else:
        print("invalid number")
```

2. Object-Oriented Implementation:

```python
class Todolist():
    #properties and methods
    #init
    def __init__(self):
        self.tasks=[]
        
    def add_task(self,task):
        self.tasks.append(task)
        
    def view_task(self):
        for i,task in enumerate (self.tasks,1):
            print(f"{i}.{task}")
            
    def mark_as_done(self,task_index):
        if 1<= task_index <=len(self.tasks):
            self.tasks[task_index-1]+="(complete)"
            
        else:
            print("Invalid task index.Please Enter a valid index value")
            
    def del_task(self,task_index):
        if 1<= task_index <= len(self.tasks):
            del self.tasks[task_index-1]
            
        else:
            print("Invalid task index .Please Enter a valid index value")
            
    
    
todolist=Todolist()

while True:
    
    print("1.add task")
    print("2.view task")
    print("3.mark as done")
    print("4.delete task")
    print("5.quit")
    
    choice=input("Enter the choice")
    
    if choice=="1":
        task=input("Enter the task:")
        todolist.add_task(task)
        
    elif choice=="2":
        todolist.view_task()
        
    elif choice=="3":
        index=int(input("Enter the task index:"))
        todolist.mark_as_done(index)
    
    elif choice=="4":
        index=int(input("Enter the task index:"))
        todolist.del_task(index)
    
    elif choice=="5":
        break;
        
    else:
        print("Invalid  number")
```
